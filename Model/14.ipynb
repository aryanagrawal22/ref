```python
import pandas as pd
import numpy as np
```

```python
df['Species'] = df['Species'].replace('Iris-setosa', 0)
df['Species'] = df['Species'].replace('Iris-versicolor', 1)
df['Species'] = df['Species'].replace('Iris-virginica', 2)
```

```python
df
```

```python
df.dtypes
```

```python
def outlier(data):
    Q1 = df[data.name].quantile(0.25)
    Q3 = df[data.name].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    upper_array = np.where(df[data.name]>=upper)[0]
    lower_array = np.where(df[data.name]<=lower)[0]
    df.drop(index=upper_array, inplace=True)
    df.drop(index=lower_array, inplace=True)
outlier(df['SepalLengthCm'])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
df['SepalLengthCm']=labelencoder.fit_transform(df['SepalLengthCm'])
df['SepalWidthCm']=labelencoder.fit_transform(df['SepalWidthCm'])
df['PetalLengthCm']=labelencoder.fit_transform(df['PetalLengthCm'])
df['PetalWidthCm']=labelencoder.fit_transform(df['PetalWidthCm'])
df.dtypes
```

```python
x = df.drop(['Species', 'Id'],axis=1)
y = df['Species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.naive_bayes import MultinomialNB
multinomial = MultinomialNB()
```

```python
multinomial.fit(x_train, y_train)
y_pred = multinomial.predict(x_test)
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))
```
