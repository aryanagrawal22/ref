```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./17_heart.csv')
df
```

```python
# Histogram

plt.figure(figsize=(12, 6))
plt.hist(df['chol'])
plt.title('Cholestrol Histogram')
plt.xlabel('Cholestrol')
plt.ylabel('People')
```

```python
# Histogram

df.hist(figsize=(10,14), bins=14)
```

```python
# Dot Plot

plt.figure(figsize=(12, 6))
plt.scatter(df['age'], df['chol'])
plt.title('Age/Chol Dot Graph')
plt.xlabel('Age')
plt.ylabel('Chol')
```

```python
# Bar Plot

plt.bar(df['age'][0:10], df['chol'][0:10], width=0.5, color='red')
```

```python
# Line Chart

plt.plot(df['age'][0:10], df['chol'][0:10])
```

```python
# Multivariate

import seaborn as sns

sns.lmplot(x='age', y='chol', hue='target', 
           data=df.loc[df['target'].isin([0,1])], 
           fit_reg=False)
```
